```python
import os
import shutil
src_dir = '/content/YOLOX/datasets/VOCdevkit/VOC2007/JPEGImages'
dst_dir = '/content/YOLOX/datasets/VOCdevkit/VOC2007/test-n'     #save directory

with open('/content/YOLOX/datasets/VOCdevkit/VOC2007/ImageSets/Main/test.txt') as f:
    
    lines = f.readlines()
    for line in lines:
        line = line.strip()
        if line != "":
            for filenames in os.listdir(src_dir):
                splitted_name = os.path.splitext(filenames)
                if (splitted_name[0] == line) and (splitted_name[1] == ".jpg"):
                    shutil.copy(os.path.join(src_dir, filenames), dst_dir)
                    print(filenames)
```

```python
import os
import shutil
src_dir = '/content/YOLOX/datasets/VOCdevkit/VOC2007/Annotations'
dst_dir = '/content/YOLOX/datasets/VOCdevkit/VOC2007/test-n'   #save directory

with open('/content/YOLOX/datasets/VOCdevkit/VOC2007/ImageSets/Main/test.txt') as f:
    lines = f.readlines()
    for line in lines:
        line = line.strip()                                                   
        if line != "":
            for filenames in os.listdir(src_dir):
                splitted_name = os.path.splitext(filenames)
                if (splitted_name[0] == line) and (splitted_name[1] == ".xml"):
                    shutil.copy(os.path.join(src_dir, filenames), dst_dir)
                    print(filenames)
```

```python
import xml.etree.ElementTree as ET

for xml_file in os.listdir('/content/YOLOX/datasets/VOCdevkit/VOC2007/test-n'):
  
   if xml_file.endswith(".xml"):
        splitted_name = os.path.splitext(xml_file)
        xml_file1 = splitted_name[0]
        p = '/content/YOLOX/datasets/VOCdevkit/VOC2007/test-txt/'+xml_file1+'.txt'   # save directory
        
        file = open(p, 'w')
        tree = ET.parse(os.path.join('/content/YOLOX/datasets/VOCdevkit/VOC2007/test-n', xml_file))
        root = tree.getroot()
        file_name = None
        for elem in root:
          
          if elem.tag == 'object':
            for subelem in elem:
                 if subelem.tag == 'name':
                      obj_name = subelem.text
                 if subelem.tag == 'bndbox':
                       xmin = ""
                       xmax = ""
                       ymin = ""
                       ymax = ""
                 for subsubelem in subelem:
                         f = float(subsubelem.text)
                         i = int(f)
                         if subsubelem.tag == "xmin":
                              xmin = str(i)
                         if subsubelem.tag == "xmax":
                                xmax = str(i)
                         if subsubelem.tag == "ymin":
                              ymin = str(i)
                         if subsubelem.tag == "ymax":
                                ymax = str(i)
                 if subelem.tag == 'bndbox':               
                     text_f = '{} {} {} {} {}'.format(obj_name, xmin, ymin, xmax, ymax)
                     file = open(p, 'a')
                     file.write(text_f)
                     file.write("\n")
        
```
